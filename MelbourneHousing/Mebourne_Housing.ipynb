### Outlier treatment ###

```python
dataset['Car'].describe()
```

```python
# Need to perform outlier analysis on variables that help in predicting the price of the house
```

```python
def getOutlier(data):
    outliers=[]
    q3=data.quantile(0.75)
    q1=data.quantile(0.25)
    iqr=q3-q1
    outliers.append(q1-1.5*iqr)
    outliers.append(q3+1.5*iqr)
    return outliers

    
    
```

```python
c1,c2=getOutlier(dataset['Car'])
```

```python
d1,d2=getOutlier(dataset['Distance'])
```

```python
l1,l2=getOutlier(dataset['Landsize'])
```

```python
r1,r2=getOutlier(dataset['Rooms'])
```

```python
def FloorandCapOutliers(data):
    for col in data.columns:
        percentiles = data[col].quantile([0.01,0.99]).values
        data[col][data[col] <= percentiles[0]] = percentiles[0]
        data[col][data[col] >= percentiles[1]] = percentiles[1]
```
